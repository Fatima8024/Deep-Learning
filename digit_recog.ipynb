```python
from keras import datasets
from keras import models
from keras import layers
from keras.layers import Flatten


(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()
train_images.shape, test_images.shape

train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255

#Create network comprising of three layers
network = models.Sequential()

network.add(Flatten(input_shape = (28,28)))
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,), name = 'layer1'))
network.add(layers.Dense(10, activation='relu', name = 'layer2'))
network.add(layers.Dense(10, activation='softmax',name = 'layer3'))

 #compile model and train
network.compile(optimizer='adam',
 loss='sparse_categorical_crossentropy',
 metrics=['accuracy'])

```

```python
 #train the model
network.fit(train_images, train_labels, epochs=10, validation_split = 0.2, batch_size=128)

 #evaluate network performance
test_loss, test_acc = network.evaluate(test_images, test_labels)

 #print the accuracy
print('test_accuracy:', test_acc)
print('test_loss:', test_loss)
```

```python
 #predictions
 y_pred = network.predict(test_images)
 y_pred_classes = np.argmax(y_pred, axis=1)

 #Classification
 print("\nClassification Report:\n")
 print(classification_report(test_labels,y_pred_classes, digits=4))
```
